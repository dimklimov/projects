# Восстановление золота из руды

Необходимо подготовить модель машинного обучения для компании «Цифры», разрабатывающей решения для эффективной работы промышленных предприятий. 
Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. 
В распоряжении имеются сырые данные, выгруженные из хранилища. Нужно использовать данные с параметрами добычи и очистки. 

Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.
Данные находятся в трёх файлах:   
- исходные данные;
- обучающая выборка;
- тестовая выборка.   

Некоторые параметры недоступны, потому что замеряются и/или рассчитываются значительно позже. Из-за этого в тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей. Также в тестовом наборе нет целевых признаков.
Исходный датасет содержит обучающую и тестовую выборки со всеми признаками.  

Необходимо:

1. Подготовить данные;
2. Провести исследовательский анализ данных;
3. Построить и обучить модель.

**Описание данных**

Технологический процесс

•	Rougher feed — исходное сырье  
•	Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant  
•	Rougher process (англ. «грубый процесс») — флотация  
•	Rougher tails — отвальные хвосты  
•	Float banks — флотационная установка  
•	Cleaner process — очистка  
•	Rougher Au — черновой концентрат золота  
•	Final Au — финальный концентрат золота  

**Параметры этапов**  
•	air amount — объём воздуха  
•	fluid levels — уровень жидкости  
•	feed size — размер гранул сырья  
•	feed rate — скорость подачи  

**Наименование признаков**  
Наименование признаков должно быть такое:  
*`этап`.`тип_параметра`.`название_параметра`*.   
Пример: rougher.input.feed_ag  

Возможные значения для блока `этап`: 

•	rougher — флотация  
•	primary_cleaner — первичная очистка  
•	secondary_cleaner — вторичная очистка  
•	final — финальные характеристики   

Возможные значения для блока `тип_параметра`:  

•	input — параметры сырья  
•	output — параметры продукта  
•	state — параметры, характеризующие текущее состояние этапа  
•	calculation — расчётные характеристики

### Проверяем правильность расчёта эффективности обогащения

Делаем это на обучающей выборке для признака rougher.output.recovery.  
Произведем рассчёты по формуле, указанной в задании, и сравним их с теми, что представлены у нас признаком в данных.  

**Recovery = (C * (F - T) / F * (C - T)) * 100%**  
*C — доля золота в концентрате после флотации/очистки;  
F — доля золота в сырье/концентрате до флотации/очистки;  
T — доля золота в отвальных хвостах после флотации/очистки.*  

Сравнивать будем с помощью параметра средней абсолютной ошибки (MAE) между расчётными и представленными значениями.    
Но сначала уберём пропуски из признака, по которому делаем расчёт, чтобы не было деление на ноль.

```python
rough_recovery_no_omissions = data_train.dropna(axis='index', how='any', subset=['rougher.output.recovery'])
```

```python
c = rough_recovery_no_omissions['rougher.output.concentrate_au'] 
f = rough_recovery_no_omissions['rougher.input.feed_au']
t = rough_recovery_no_omissions['rougher.output.tail_au']

recovery_output = rough_recovery_no_omissions['rougher.output.recovery']

def recovery_formula(recovery, C, F, T):
    recovery_form = pd.Series((C * (F - T)) / (F * (C - T)) * 100)
    mae = mean_absolute_error(recovery_form, recovery).round(16)
    print('Средняя абсолютная ошибка (MAE) между расчётными и указанными данными составляет', mae)
    
recovery_formula(recovery_output, c, f, t)
```

МАЕ между значениями ничтожно мало и составляет не более 9.4e-15. Можем сделать вывод, что представленные в признаке `rougher.output.recovery` значения  рассчитаны верно.

```python
Au = data_full[['rougher.input.feed_au','rougher.output.concentrate_au','primary_cleaner.output.concentrate_au','final.output.concentrate_au']]
Ag = data_full[['rougher.input.feed_ag','rougher.output.concentrate_ag','primary_cleaner.output.concentrate_ag','final.output.concentrate_ag']]
Pb = data_full[['rougher.input.feed_pb','rougher.output.concentrate_pb','primary_cleaner.output.concentrate_pb','final.output.concentrate_pb']]

def rename_columts(df):
    df.columns = ['rougher.input','rougher.output', 'primary_cleaner.output', 'final.output']
    
for df in Au, Ag, Pb:
    rename_columts(df)
```

```python
def concentration_analyst(df, title):
    sns.set(style='whitegrid')
    fig, ax = plt.subplots(1,4, sharey=True, figsize=(10,5))
    columns = df.columns
    colors = ['brown', 'red', 'orange', 'yellow']
    for i in range(len(columns)):
        sns.boxplot(color=colors[i], y=columns[i], linewidth=2, ax=ax[i], data=df)
        ax[i].set_xlabel(columns[i])
        ax[i].set_ylabel('')
    fig.suptitle('Концентрация {} на различных этапах очистки'.format(title), size=14)
    fig.show()
    
for df in Au, Ag, Pb:
    concentration_analyst(df, get_title(df))
```

## Модель

### Функция для вычисления итоговой sMAPE

Пишем функцию для sMAPE согласно формуле в задании.

```python
def smape(target, predictions):
    numerator = abs(predictions - target)
    denominator = (abs(target) + abs(predictions)) / 2
    summ = (numerator / denominator).sum()

    smape = (1 / len(target)) * summ * 100
    return smape
```

Итоговое sMAPE считаем по 2м величинам.

```python
def smape_final(rougher, final):
    smapefinal = 0.25 * rougher + 0.75 * final
    return smapefinal.round(2)
```

### Обучение разных моделей и оценка их качества

Оставим в трейне только те, признаки, что есть в тесте.  
И тут я горько осознал, что нулевую конценртацию нужно было только в сырье удалять, так как сейчас всё лишнее снесём.

Разобьём выборки на фичи и таргеты

```python
def features_target_split(train, test):

    features_train = train.drop(['rougher.output.recovery', 'final.output.recovery', 'date'], axis=1)
    target_train = train[['rougher.output.recovery', 'final.output.recovery']]

    
    features_test = test.drop(['rougher.output.recovery', 'final.output.recovery', 'date'], axis=1)
    target_test = test[['rougher.output.recovery','final.output.recovery']]
        
    return features_train, target_train, features_test, target_test

features_train, target_train, features_test, target_test = features_target_split(data_train, data_test)
```

Масштабируем признаки с помощью стандартизации данных, чтобы модели равнозначно учитывали все признаки.

```python
scaler = StandardScaler()
scaler.fit(features_train[features_train.columns]) 
features_train[features_train.columns] = scaler.transform(features_train[features_train.columns])
features_test[features_train.columns]= scaler.transform(features_test[features_train.columns]) 
features_train.head()
```

Применим GridSearchCV для отбора модели. По сути та же кросс-валидация, только с сохраненнием лучших параметров.

```python
def grid_search_cv(features, target):

    model_and_parameters = [
                            (Lasso(random_state=12345), {'alpha' : np.linspace(0.1, 3, 20)}),
                            (Ridge(random_state=12345), {'alpha' : np.linspace(0.1, 3, 20)}),
                            (LinearRegression(), {}),
                            (RandomForestRegressor(random_state=12345), {'n_estimators': range(1,101,10), 'max_depth': range(5,55,5)})
                            ]
    
    best_final_smape = 1000

    for model, params in model_and_parameters:

        scorer = make_scorer(smape, greater_is_better=False)
        grid_rougher = GridSearchCV(estimator=model, param_grid=params, cv=5, scoring=scorer, n_jobs=-1)
        grid_final = GridSearchCV(estimator=model, param_grid=params, cv=5, scoring=scorer, n_jobs=-1)

        grid_rougher.fit(features, target['rougher.output.recovery'])
        rougher_smape = grid_rougher.best_score_

        grid_final.fit(features, target['final.output.recovery'])
        final_smape = grid_final.best_score_

        value_final_smape = smape_final(rougher_smape, final_smape)

        if value_final_smape < best_final_smape:
            best_model_rougher = grid_rougher.best_estimator_
            best_model_final = grid_final.best_estimator_
            best_final_smape = value_final_smape

        print(model.__class__.__name__, ":", value_final_smape)
        
    return best_model_rougher, best_model_final
    
best_model_rougher, best_model_final = grid_search_cv(features_train, target_train)
```

Лучше всего на  трейне  себя проявляет модель Lasso.

Отправляем на тест

```python
pred_rough = best_model_rougher.predict(features_test)
pred_final = best_model_final.predict(features_test)
smape_final(smape(pred_rough, target_test['rougher.output.recovery']), 
           smape(pred_final, target_test['final.output.recovery']))
```

Проверим её адекватность, сравнив с константной моделью по медиане.

```python
model_median = DummyRegressor(strategy="median")

model_median.fit(features_train, target_train['rougher.output.recovery'])
pred_rough = model_median.predict(features_test)
model_median.fit(features_train, target_train['final.output.recovery'])
pred_final = model_median.predict(features_test)

print("Прогноз констанной модели по медиане: {:.2f}".format(smape_final(smape(pred_rough, target_test['rougher.output.recovery']), 
           smape(pred_final, target_test['final.output.recovery']))))
```

Лучшая модель ошибается реже константной, что говорит о её адекватности.
